Usando FastAPI

```python
from fastapi import FastAPI
app = FastAPI(
    title=" API in Google Colab",
    description="This is a practice API with FastAPI",
    version="1.0.0",
    docs_url="/",
    redoc_url="/redoc"
)

@app.get("/welcome/")
def read_root():
    return {"mensaje: Hola, mundo!"}

@app.get("/items/{item_id}/")
def read_item(item_id: int, q: str = None):
    return {"item_id": item_id, "q": q}
```

Ejecutar el servidor

```python
from pyngrok import ngrok
import uvicorn
import threading

# Crear túnel con ngrok
public_url = ngrok.connect(8000)
print(f"Public URL: {public_url}")

# Ejecutar FastAPI en segundo plano
def run():
    uvicorn.run(app, host="0.0.0.0", port=8000)

thread = threading.Thread(target=run)
thread.start()
```

Detener el servidor

```python
ngrok.disconnect(public_url)
print("Tunel ngrok desconectado")
ngrok.kill()
print("Todos los túneles ngrok han sido cerrados")
```
